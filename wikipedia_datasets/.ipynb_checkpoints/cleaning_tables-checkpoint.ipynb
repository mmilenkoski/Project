## Cleaning the wikipedia tables of the election results
In this notebook we clean the manually scrapped tabels for the last and the previous elections of the countries of intereset. First, we read the notebooks, we select only the relevant columns, we compute the percent of seats every party has in parlament and we save these claned data frames in separe folders.

```python
import glob
import os
import pandas as pd
import numpy as np
```

```python
# Directories from where to read and where to store the tables
last_elections_dir = "last_election/"
second_to_last_election_dir = "second_to_last_election/"

last_elections_claned_dir = "last_election_cleaned/"
second_to_last_claned_election_dir = "second_to_last_election_cleaned/"
```

```python
# Get all tables for every country for the last and previous elections
last_elections = glob.glob(last_elections_dir + "*.csv")
second_to_last_election = glob.glob(second_to_last_election_dir + "*csv")

last_elections = sorted(last_elections)
second_to_last_election = sorted(second_to_last_election)
```

```python
# For every country get the last and the previous elections
for election1_path, election2_path in zip(last_elections, second_to_last_election):
    election1_df = pd.read_csv(election1_path)
    election2_df = pd.read_csv(election2_path)
    
    # Select only the relevant columns
    election1_df = election1_df[["Party", "Votes", "Votes %", "Seats"]]
    election2_df = election2_df[["Party", "Votes", "Votes %", "Seats"]]
    
        
    # Compute the percent of seats
    e1_seats = np.sum(election1_df["Seats"])
    e2_seats = np.sum(election2_df["Seats"])
    e1_seats_percent = (election1_df["Seats"] / e1_seats) * 100
    e2_seats_percent = (election2_df["Seats"] / e2_seats) * 100
    election1_df["Seats %"] = e1_seats_percent
    election2_df["Seats %"] = e2_seats_percent
    
    # Save the claned data frames as csv files
    path1, file1 = os.path.split(election1_path)
    path2, file2 = os.path.split(election2_path)
    
    print("Saving: %s" % file1)
    election1_df.to_csv(os.path.join(last_elections_claned_dir, file1))
    print("Saving: %s" % file2)
    election2_df.to_csv(os.path.join(second_to_last_claned_election_dir, file2))
    print()
```
