## Generating the final dataset
In this notebook we merge the results of the last and previous elections for every country of intereset in one final table. Also, for every party we add the political position ("left", "right", "center", etc.) in the final dataset.

```python
import glob
import os
import pandas as pd
import numpy as np
```

```python
# Directories from where to read the cleaned tables
last_elections_dir = "last_election_cleaned/"
second_to_last_election_dir = "second_to_last_election_cleaned/"
```

```python
# Get all tables for every country for the last and previous elections
last_elections = glob.glob(last_elections_dir + "*.csv")
second_to_last_election = glob.glob(second_to_last_election_dir + "*csv")

last_elections = sorted(last_elections)
second_to_last_election = sorted(second_to_last_election)
```

```python
# Data frame where to store the final table
all_countries_df = pd.DataFrame()

# For every country get the last and the previous elections
for election1_path, election2_path in zip(last_elections, second_to_last_election):
    # Extract the country name from the table file names
    file, path = os.path.split(election1_path)
    country = path.split("_")[0]
    
    df1 = pd.read_csv(election1_path, index_col=0)
    df2 = pd.read_csv(election2_path, index_col=0)
    
    # Merge the elections tables for a given country
    df_merged = pd.merge(df1, df2, on="Party", suffixes=["_e1", "_e2"])
    df_merged.insert(0, "Country", country)
    
    # Add the merged tables to the final table
    all_countries_df = pd.concat([all_countries_df, df_merged])
        
all_countries_df.index = range(0, all_countries_df.shape[0])
```

```python
# Save the final table
all_countries_df.to_csv("all_countries_df.csv")
```

```python
# Read the table containing the political position of every European party
parties_df = pd.read_csv("all_parties.csv")
# Cleaning the names of the political position
ps = np.array(parties_df["Political_position"])
parties_df["Political_position"] = np.array([x.lower() for x in ps])
```
