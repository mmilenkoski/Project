```python
# imports
import pandas as pd
import numpy as np

import findspark
findspark.init()
from pyspark import SparkContext
from pyspark.sql import *
from pyspark.sql.types import StringType
from pyspark.sql.functions import *

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# initialize spark context
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
df = spark.read.parquet("data_frame.parquet")
```

```python
df.count()
```

```python
df.show()
```

```python
df.select("Day_DATE").take(1000)
```

```python
#avg_tone = df.groupBy("Day_DATE").avg("AvgTone")
avg_tone = df.groupBy("MonthYear_Date").avg("GoldsteinScale")
```

```python
avg_tone_pd = avg_tone.toPandas()
```

```python
avg_tone_pd.columns = ["day", "avgTone"]
```

```python
avg_tone_pd = avg_tone_pd.sort_values("day")
```

```python
avg_tone_pd.plot(x="day", y="avgTone")
```

```python
# Plot the results
plt.figure(figsize=(14, 8))
ax = sns.lineplot(x="day", y="avgTone", data=avg_tone_pd, color="lightskyblue")
#start, end = plot_df["index"].min(), plot_df["index"].max()
#tics = range(start, end, 4)
#plt.xticks(tics, np.array(plot_df["dates"])[tics], rotation=45)
plt.xlabel("Date")
plt.ylabel("Number of PySpark questions")
plt.plot()
```
